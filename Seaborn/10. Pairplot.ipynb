```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
```

# [Seaborn Pairplot](https://seaborn.pydata.org/generated/seaborn.pairplot.html)
Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each variable in data will by shared in the y-axis across a single row and in the x-axis across a single column. The diagonal Axes are treated differently, drawing a plot to show the univariate distribution of the data for the variable in that column.


```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
g = sns.pairplot(iris);
```

## Show different levels of a categorical variable by the color of plot elements:

```python
g = sns.pairplot(iris, hue="species");
```

## Use a different color palette:

```python
sns.pairplot(iris, hue="species", palette="husl");
```

## Use different markers for each level of the hue variable:

```python
g = sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
```

## Plot a subset of variables using the variable - `vars`:

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"]);
```

## Draw larger plots:

```python
sns.pairplot(iris, height=3, vars=["sepal_width", "sepal_length"]);
```

## Plot different variables in the rows and columns:

```python
sns.pairplot(iris, x_vars = ["sepal_width", "sepal_length"], y_vars = ["petal_width", "petal_length"]);
```

## Use kernel density estimates for univariate plots:

```python
sns.pairplot(iris, diag_kind="kde");
```

## Fit linear regression models to the scatter plots using `kind = reg`:

```python
sns.pairplot(iris, kind="reg");
```

## Pass keyword arguments down to the underlying functions 

```python
sns.pairplot(iris, diag_kind = 'kde', markers = '+', diag_kws=dict(shade=True),
             plot_kws=dict(s=50, edgecolor="b", linewidth=1));
```
