# Seaborn Introduction

[Seaborn](https://seaborn.pydata.org/) is a Python data visualization library based on [matplotlib](https://matplotlib.org/). It provides a high-level interface for drawing attractive and informative statistical graphics.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

The below figure shows the default matplotlib theme.

```python
x = np.random.randint(1, 15, 20)
y = np.random.randint(1, 15, 20)
plt.scatter(x,y);
plt.xlabel('x', fontsize = 15);
plt.ylabel('y', fontsize = 15);
plt.title('Scatterplot of x and y', fontsize = 15);
```

## Applying seaborn styles to all figures

The style theme provided by seaborn is better than that of the matplotlib library.

The default theme of seaborn is called '__darkgrid__'.

To get the default theme for all figures use `sns.set()` once after importing the seaborn and matplotlib library.

```python
sns.set()
```

```python
plt.scatter(x,y);
plt.xlabel('x', fontsize = 15);
plt.ylabel('y', fontsize = 15);
plt.title('Scatterplot of x and y', fontsize = 15);
```

We can change the theme using __sns.set_style()__. 

It takes one of five arguments/theme-names: __white, dark, whitegrid, darkgrid, and ticks__. These are the five options for the background of your plot; the default one is __darkgrid__.

##### Themes can be set using the function `sns.set_style(theme-name)`:

> sns.set_style('whitegrid')


```python
sns.set_style('whitegrid')
plt.scatter(x,y);
plt.xlabel('x', fontsize = 15);
plt.ylabel('y', fontsize = 15);
plt.title('Scatterplot of x and y', fontsize = 15);
```

## Setting the figure size

In matplotlib, we have to use the following command to adjust the size of the image: `plt.figure(figsize=(width, height))`. 

set figure size by passing dictionary to rc parameter with key figure.figsize in seaborn set method.  
Eg: `sns.set(rc={'figure.figsize':(11,9))`

```python
sns.set(rc={'figure.figsize':(11,9)})
```

```python
sns.set_style('whitegrid')
plt.scatter(x,y);
plt.xlabel('x', fontsize = 15);
plt.ylabel('y', fontsize = 15);
plt.title('Scatterplot of x and y', fontsize = 15);
```

## Seaborn color palettes

### Seaborn default color palette

Seaborn has six variations of its default color palette: `deep`, `muted`, `pastel`, `bright`, `dark`, and `colorblind`.

![](palettes.PNG)

To use one of these palettes, pass the name into `sns.set_palette()`:

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.set_palette('colorblind');
sns.boxplot(data = iris);
```

### USING COLOR BREWER PALETTES

Seaborn also allows the use of Color Brewer palettes. 

![](palettes_2.PNG)

***
__Qualitative Palettes:__  When using a dataset that uses distinct but non-ordered categories, it’s good to use qualitative palettes. Qualitative palettes are sets of distinct colors which make it easy to distinguish the categories when plotted but don’t imply any particular ordering or meaning.

__SEQUENTIAL PALETTES:__ sequential palettes are a set of colors that move sequentially from a lighter to a darker color. Sequential color palettes are appropriate when a variable exists as ordered categories, such as grade in school, or as continuous values that can be put into groups, such as yearly income. Because the darkest colors will attract the most visual attention, sequential palettes are most useful when only high values need to be emphasized.

__Diverging palettes:__ are best suited for datasets where both the low and high values might be of equal interest, such as hot and cold temperatures.
