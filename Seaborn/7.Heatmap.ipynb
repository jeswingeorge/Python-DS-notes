***

Plot a heatmap for a numpy array:

````python
uniform_data = np.random.rand(10, 12);
plt.figure(figsize=(10,6))
sns.heatmap(uniform_data);
````

#### Change the limits of the colormap

````python
plt.figure(figsize=(10,6))
sns.heatmap(uniform_data, vmin=0, vmax=1);
````

#### Plot a heatmap for data centered on 0 with a diverging colormap:

````python
normal_data = np.random.randn(10, 15)
plt.figure(figsize=(10,6))
sns.heatmap(normal_data, center=0);
````

````python
flights = sns.load_dataset("flights")
flights.head()
````

````python
flights = flights.pivot(index = "month", columns = "year", values = "passengers")
flights
````

````python
plt.figure(figsize=(10,5))
sns.heatmap(flights);
````

#### Annotate each cell with the numeric value using integer formatting:
Use parameters: `annot=True` - If True, write the data value in each cell.
and `fmt` - String formatting code to use when adding annotations. 

By default `fmt` is set to scientific notations for numbers i.e.,__3.5e02__ 
Formats: 
`'d'` - integers 
`'f'` - floating point numbers


````python
plt.figure(figsize=(12,6))
sns.heatmap(flights, annot=True, fmt = 'd');
````

#### Add lines between each cell using parameter `linewidth`:

````python
sns.heatmap(flights, linewidths=.5);
````

Heat map works by correlation. This shows you which variables are correlated to each other from a scale of 1 being 
the most correlated and -1 is not correlated at all. However, you cannot correlate strings. You can only correlate 
numerical features. Dont let this stop you because you can always encode your categorical data types using the get_dummies 
function in Pandas library.


````python
iris = sns.load_dataset('iris')
iris.head()
````

````python
plt.figure(figsize=(10,6))
sns.heatmap(iris.corr());
````

#### Converting categorical varaible (species) to dummy variables

````python
dataset = pd.get_dummies(iris, columns = ['species'], drop_first = True)
dataset.head()
````

````python
plt.figure(figsize=(10,6))
sns.heatmap(dataset.corr(), annot = True);
````

#### Change the Heatmap Colors

````python
plt.figure(figsize=(9,5))
sns.heatmap(iris.corr(), annot=True, linewidth = 0.5, cmap='coolwarm');
````
