```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set()
sns.set(rc={'figure.figsize':(10,8)})
```

We can develop a QQ plot in Python using the [qqplot() statsmodels function](https://www.statsmodels.org/dev/generated/statsmodels.graphics.gofplots.qqplot.html). The function takes the data sample and by default assumes we are comparing it to a Gaussian distribution. We can draw the standardized line by setting the ‘line‘ argument to ‘s‘.

```python
from statsmodels.graphics.gofplots import qqplot
```

```python
data = 7 * np.random.randn(1000) + 180
# q-q plot
qqplot(data, line='s');  # Always use line-s
```

Alternatively, we can use `stats.probplot()` from the scipy library.

```python
import scipy.stats as stats
```

```python
stats.probplot(data, dist="norm", plot=plt)
plt.title("Normal Q-Q Plot")
```
