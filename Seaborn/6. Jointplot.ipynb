```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
sns.set(style = 'ticks')

# pandas defaults
pd.options.display.max_columns = 500
pd.options.display.max_rows = 500
```

## [Seaborn Jointplot](https://seaborn.pydata.org/generated/seaborn.jointplot.html)

- Can be used for visualizing multivariate distributions.
- It can be used to draw a plot of two variables with bivariate and univariate graphs.


__Reference:__ [Simple ways to improve using matplotlib](https://towardsdatascience.com/simple-ways-to-improve-your-matplotlib-b64eebccfd5)


```python
tips = sns.load_dataset('tips')
tips.head()
```

#### Draw a scatterplot with marginal histograms:

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip');
```

#### Add regression and kernel density fits:

Parameter `kind : { “scatter” | “reg” | “resid” | “kde” | “hex” }` to show the kind of plot that is required. 

- `kind = 'scatter'` : gives a scatterplot (this is used by default).
- `kind = 'reg'` : gives a regression line along with the scatterplot
- `kind = 'resid'` : gives a residual plot.
- `kind = 'kde'` : Replaces the histogram with a KDE Plot described as Kernel Density Estimate (used for visualizing the Probability Density of a continuous variable) and  the scatterplot with contours.
- `kind = 'hex'` : replaces scatterplot with a joint histogram using hexagonal bins

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'reg');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'resid');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'kde');
```

```python
sns.jointplot(data = tips, x = 'total_bill', y = 'tip', kind = 'hex');
```

#### align the marginal Axes tightly with the joint Axes:

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.jointplot(data = iris, x = 'petal_length', y = 'sepal_length', kind = 'kde', space = 0, color = 'b');
```


#### Draw a scatterplot, then add a joint density estimate using `plot_joint` method:

Using `plot_joint` method we can add an additional plot to the scatterplot of the joint plot.

```python
# n_levels - number of contour levels
(sns.jointplot(data = iris, x = 'petal_length', y = 'sepal_length', color = 'k').plot_joint(sns.kdeplot, n_levels = 6));
```

 #### Pass vectors in directly without using Pandas, then name the axes:

```python
x, y = np.random.randn(2, 300)
sns.jointplot(x,y, kind = 'hex').set_axis_labels("x", "y");
```

#### Draw a smaller figure with more space devoted to the marginal plots

- `height`(numeric) - Size of the figure (it will be square).
- `ratio`(numeric) - Ratio of joint axes height to marginal axes height. (Smaller the ratio more the height of marginal figure).
- `{joint, marginal, annot}_kws : dicts, optional` - Additional keyword arguments for the plot components.

```python
sns.jointplot(data = iris, x = 'petal_length', y = 'sepal_length', color = 'b', height=9, ratio=2);
```

#### Pass keyword arguments down to the underlying plots:

```python
sns.jointplot(data = iris, x = 'petal_length', y = 'sepal_length',
              joint_kws = dict(color = 'g'),marginal_kws=dict(bins=10, rug=True),s=40, edgecolor="k", linewidth=1);
```
