## [Seaborn Lineplot](https://seaborn.pydata.org/generated/seaborn.lineplot.html)

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(style="ticks", rc={'figure.figsize':(8,7)})
sns.set_context(rc = {"font.size":15, "axes.labelsize":15}, font_scale=2)
sns.set_palette('colorblind');
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

#### Draw a single line plot without showing a confidence interval:

```python
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', ci = None);
```

#### Draw a single line plot with error bands showing a confidence interval:

```python
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal');
```

#### Group by another variable and show the groups with different colors:

```python
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', ci = None, hue = 'event');
```

#### Show the grouping variable with both color and line dashing:

```python
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', ci = None, hue = 'event', style = 'region');
```

#### Show error bars instead of error bands and plot the standard error:

```python
sns.lineplot(data = fmri, x = 'timepoint', y = 'signal', err_style="bars", ci = 68, hue = 'event');
```

#### Plot from a wide-form DataFrame:

```python
index = pd.date_range("1 1 2000", periods=100,freq="m", name="date")
data = np.random.randn(100, 4).cumsum(axis=0)
wide_df = pd.DataFrame(data, index, ["a", "b", "c", "d"])
wide_df.head()
```

```python
sns.lineplot(data=wide_df);
```

Use [relplot()](https://seaborn.pydata.org/generated/seaborn.relplot.html#seaborn.relplot) to combine `lineplot(FacetGrid)` and `FacetGrid`: This allows grouping within additional categorical variables. Using `relplot()` is safer than using FacetGrid directly, as it ensures synchronization of the semantic mappings across facets.

```python
sns.relplot(data = fmri, x = 'timepoint', y = 'signal', hue = 'event', col = 'region', kind = 'line');
```
