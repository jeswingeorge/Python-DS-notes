# [Pandas Dataframe Rank](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.rank.html)

Compute numerical data ranks (1 through n) along axis.

By default, equal values are assigned a rank that is the average of the ranks of those values. We can change this using the parameter `method = ‘max’` and `ascending = False`.

```python
df = pd.read_excel("..\data\superstore_sample.xlsx")
df.head()
```

```python
temp =  pd.pivot_table(data = df, values = 'Sales', index = 'Product Sub-Category', columns = 'Ship Mode', aggfunc = np.sum,
                      fill_value=0)
temp
```

## Find the rank of __Product Sub-category's__ sales  based on  each of the __Ship Mode__


```python
temp['Regular Air Rank'] = temp['Regular Air'].rank(method='max', axis = 'index', ascending = False)
temp['Delivery Truck Rank'] = temp['Delivery Truck'].rank(method='max', axis = 'index',ascending = False)
temp['Express Air Rank'] = temp['Express Air'].rank(method='max', axis = 'index', ascending = False)
```
