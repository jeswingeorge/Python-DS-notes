```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame()
df['A'] = np.random.randint(40, 100, size = 7)
df['B'] = np.random.randint(40, 100, size = 7)
df['C'] = np.random.randint(40, 100, size = 7)
df['D'] = np.random.randint(40, 100, size = 7)
df
```

## 1. Using Apply function on a particular row/column of dataframe

#### Eg 1: To  add 10 to each element of column B

```python
df['B'] = df['B'].apply(lambda x: x+10)
df
```

#### Eg 2: To add 100 to each element of index 3

```python
df.iloc[3,:] = df.iloc[3,:].apply(lambda x: x + 100)
df
```

## 2. Using apply function on a dataframe rowwise/columnwise

```python
student = pd.DataFrame()
student['Name'] = list('ABCDEFG')
student['Physics'] = np.random.randint(40, 100, size = 7)
student['Chemistry'] = np.random.randint(40, 100, size = 7)
student['Biology'] = np.random.randint(40, 100, size = 7)
student['English'] = np.random.randint(40, 100, size = 7)
student
```

#### 1. To find the maximum marks scored by each student

```python
student['Maximum mark'] = student.loc[:, 'Physics':'English'].apply(max, axis = 'columns')
student
```

#### 2. To find the total marks scored by each student

```python
student['Total Mark'] = student.loc[:, 'Physics':'English'].apply(sum, axis = 'columns')
student
```

#### 3. To create a dictionary for each student having key as _Maximum mark_ and _Total Mark_

```python
def student_dict_creation(row):
    student_dict = {}
    student_dict['Maximum mark'] = row['Maximum mark']
    student_dict['Total mark'] = row['Total Mark']
    return(student_dict)
```

```python
student['student_dict'] = student.apply(student_dict_creation, axis = 'columns')
student
```
