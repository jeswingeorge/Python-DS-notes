```python
import pandas as pd
import numpy as np
```

## Case 1. sort a group within the first group by using a numerical value

```python
df = pd.DataFrame()
df['Class'] = [10, 10, 9, 8, 10, 7, 9, 8, 7, 10, 8, 10, 9, 7, 8, 10]
df['Student'] = [chr(var) for var in range(ord('a'), (ord('a')+len(df.Class)))]
df['Marks'] = np.random.randint(low = 79, high = 100, size = len(df.Class))
df
```

```python
# Group by class and arrange in descending order the students for each class based on their marks
tt = df.groupby(by = ['Class'], as_index = False).apply(lambda gd: gd.sort_values(by = 'Marks', ascending = False))
tt.head(10)
```

```python
tt.index
```

[Pandas multilevel index](https://www.geeksforgeeks.org/python-pandas-multiindex-droplevel/)

```python
tt.droplevel(level = 1)
```

```python
tt.index = [i for i in range(len(tt))]
tt
```

## Case 2. based on the count of categorical variable in a group sort each group

```python
market = pd.DataFrame()
store_t = ['Amazon', 'D-Mart', 'R-Trends', 'Flipkart', 'Big Bazar']
market["store"] = [store_t[i] for i in np.random.randint(low = 0, high = len(store_t), size = 500)]
items = ['Dettol', 'Rin Soap', 'Sandisk USB', 'Harpic', 'Maggi', 'Kellogs', 'Munch', 'Perk', 'Jim Jam', 'Xiomi Redmi 2', \
         'Oneplus', 'Kit-Kat', 'Kurkure', 'Lays']
market["items"] = [items[i] for i in np.random.randint(low = 0, high = len(items), size = 500)]
market.head()
```

```python
market.groupby(by = ['store']).apply(lambda gd: gd['items'].value_counts().sort_values(ascending = False))
```

## Case 3. Based on sum of numerical variable sort the second-level group within thefirst level group

```python
football = pd.DataFrame()
team = ['Barcelona', 'Manchester United', 'Chelsea', 'Real Madrid', 'PSG', 'Juventus', 'AC Milan']
football['team'] = [team[i] for i in np.random.randint(low = 0, high = len(team), size = 50)]
football['player_number'] = np.random.randint(low = 3, high = 13, size = 50)
football['goals_scored'] = np.random.randint(low = 0, high = 10, size = 50)
football.head()
```

```python
(football.groupby(['team', 'player_number']).sum().reset_index().groupby('team', as_index = False)
 .apply(lambda gd:gd.sort_values('goals_scored', ascending = False))).reset_index().drop(['level_0', 'level_1'], axis = 1)
```
