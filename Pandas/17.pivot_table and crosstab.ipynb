# Pivot table
A pivot table is a data summarization tool that is used in the context of data processing. Pivot tables are used to summarize, sort, reorganize, group, count, total or average data stored in a database. It allows its users to transform columns into rows and rows into columns. It allows grouping by any data field. Pivot tables are the perfect solution when you need to summarize and analyze large amounts of data. In just a few clicks, you have access to a whole new set of information.

```python
df = pd.read_csv('../data/titanic-data.csv')
```

```python
df.head()
```

[Pandas pivot_table()](https://www.geeksforgeeks.org/python-pandas-pivot_table/)

Average age of passengers in Pclass based on their survival

```python
pd.pivot_table(data = df, values = 'Age',index = 'Pclass', columns = 'Survived', aggfunc = np.mean)
```

Average age of passengers in Pclass based on their survival and gender

```python
pd.pivot_table(data = df, values = 'Age',index = 'Pclass', columns = ['Survived', 'Sex'], aggfunc = np.mean)
```

Total fare paid by passengers based on the passenger class and port of embarkment and their gender.

Use `fill_value = 0` to replace 0 with Null values.

```python
pd.pivot_table(data = df, values = 'Fare',index = ['Embarked', 'Pclass'], columns = ['Survived', 'Sex'], aggfunc = np.sum,
              fill_value = 0)
```

To find total the row and columns sub total and grand total as in Tableau

```python
pd.pivot_table(data = df, values = 'Fare',index = ['Embarked', 'Pclass'], columns = ['Survived', 'Sex'], aggfunc = np.sum,
              fill_value = 0, margins = True)
```

## [Pandas Crosstab](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.crosstab.html)

Compute a simple cross tabulation of two (or more) factors. By default computes a frequency table of the factors unless an array of values and an aggregation function are passed.

```python
pd.crosstab(index = df['Pclass'], columns = df['Survived'])
```

```python
pd.crosstab(index = [df['Embarked'], df['Pclass']], columns = df['Survived'])
```

To calculate the subtotals (as in Tableau) use `margins = True`

```python
pd.crosstab(index = [df['Embarked'], df['Pclass']], columns = df['Survived'], margins = True)
```

To find the percentage of total use: __Normalize__ by dividing all values by the sum of values.

- If passed ‘all’ or True, will normalize over all values.
- If passed ‘index’ will normalize over each row.
- If passed ‘columns’ will normalize over each column.
- If margins is True, will also normalize margin values.

```python
pd.crosstab(index = [df['Embarked'], df['Pclass']], columns = df['Survived'], margins = True, normalize = 'all')
```
