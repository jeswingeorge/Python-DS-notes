## [Merge 2 dataframes](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.merge.html)

```python
import pandas as pd
import numpy as np
pd.__version__
```

```python
student_dict = {
    "Roll number": [1,2,3,4,5],
    "Student": ['A', 'B', 'C', 'D', 'E'],
    "Age": [17, 16, 18, 16, 15]
}

student_df = pd.DataFrame(student_dict)
student_df
```

```python
marks_df = pd.DataFrame({
    "Roll no.": [2,4,6,8,9,10,5],
    "Marks": [90, 89, 78, 87, 70, 93, 96]
})
marks_df
```

### Inner join

```python
student_df.merge(marks_df, left_on = 'Roll number', right_on = 'Roll no.', how = 'inner')
```

### left_join

```python
student_df.merge(marks_df, left_on = 'Roll number', right_on = 'Roll no.', how = 'left')
```

### right_join

```python
student_df.merge(marks_df, left_on = 'Roll number', right_on = 'Roll no.', how = 'right')
```

### outer join

```python
student_df.merge(marks_df, left_on = 'Roll number', right_on = 'Roll no.', how = 'outer')
```
