```python
import pandas as pd
import numpy as np
```

### Reason to Cut and Bin your Continous Data into Categories

1. Wide range of numerical data that will be more readable in groups
2. Need for statistical analysis of groups for better insight

```python
df = pd.DataFrame()
df['Age'] = np.random.randint(5, 50, size = 20)
df
```

### Create New Column of age_bins Via Defining Bin Edges

```python
df['age_bins'] = pd.cut(df['Age'], bins = [0,10,20,30,40,50])
df
```

```python
df['age_bins'].unique()
```

### Adding labels to the categories

```python
df['age_categories'] = pd.cut(df['Age'], bins = [0,10,20,30,40,50], labels = ['child','10s', '20s', '30s', '40s'])
df
```
