````python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
````

## Finding the coordinates where two curves intersect

### Example - 1

````python
# x-axis values
x = np.arange(0,1000)
x[:10]
````

````python
# 1st curve
f = np.arange(0,1000)
f[:10]
````

````python
# 2nd curve - sine wave
g = np.sin(np.arange(0, 10, 0.01) * 2) * 1000
g[:10]
````

````python
plt.figure(figsize = (8,6));
plt.plot(x, f, '-', label = 'Line');
plt.plot(x, g, '-', label = 'Sine wave');
plt.legend();
idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
plt.plot(x[idx], f[idx], 'ro');
plt.show();
````

First it calculates `f - g` and the corresponding signs using `np.sign`. Applying `np.diff` reveals all the positions, where the sign changes (e.g. the lines cross). Using `np.argwhere` gives us the exact indices.

We get all the intersection points using the `np.argwhere(np.diff(np.sign(f - g))).flatten()` command.

````python
idx = np.argwhere(np.diff(np.sign(f - g))).flatten()
idx
````

To get the first and third intersection point we can use indexing.

````python
print('x-axis values of the first and third intersection points: ')
x[idx[0]], x[idx[2]]
````

````python
print('y-axis values of the first and third intersection points: ')
g[idx[0]], g[idx[2]]
````

#### Reference:
1. [np.sign](https://numpy.org/doc/stable/reference/generated/numpy.sign.html)
    
    Returns an element-wise indication of the sign of a number.  
    The sign function returns -1 if x < 0, 0 if x==0, 1 if x > 0. nan is returned for nan inputs.
    
    ```
    >>> np.sign([-5., 4.5])
        array([-1.,  1.])
    >>> np.sign(0)
        0
    >>> np.sign(5-2j)
        (1+0j)
    ```

2. [np.diff](https://numpy.org/doc/stable/reference/generated/numpy.diff.html)

    Calculate the n-th discrete difference along the given axis.

    The first difference is given by `out[i] = a[i+1] - a[i]` along the given axis, higher differences are calculated by using diff recursively.
